```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')
```

```python
file_path = '../data/raw/healthcare-dataset-stroke-data.csv'

df = pd.read_csv(file_path)
```

```python
print("5 dòng dữ liệu đầu tiên:")
display(df.head())

print("\nThông tin tổng quan (df.info()):")
df.info()

print("\nThống kê mô tả cho các cột số (df.describe()):")
display(df.describe())

print(f"\nKích thước của dataset: {df.shape[0]} dòng và {df.shape[1]} cột")
```

```python
stroke_counts = df['stroke'].value_counts()
print("Phân phối của biến mục tiêu 'stroke':")
print(stroke_counts)

plt.figure(figsize=(8, 6))
sns.countplot(x='stroke', data=df)
plt.title('Phân phối số ca bị đột quỵ (1) và không bị đột quỵ (0)', fontsize=16)
plt.ylabel('Số lượng')
plt.xlabel('Đột quỵ')
plt.xticks([0, 1], ['Không', 'Có'])
plt.show()
```

```python

categorical_features = ['gender', 'hypertension', 'heart_disease', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']

for col in categorical_features:
    plt.figure(figsize=(10, 5))
    sns.countplot(x=col, data=df)
    plt.title(f'Phân phối của {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
numerical_features = ['age', 'avg_glucose_level', 'bmi']

for col in numerical_features:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Phân phối của {col}')
    plt.show()
```

```python

for col in numerical_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='stroke', y=col, data=df)
    plt.title(f'{col} vs. Stroke')
    plt.xticks([0, 1], ['Không', 'Có'])
    plt.show()
```

```python
for col in categorical_features:
    plt.figure(figsize=(12, 6))
    sns.countplot(x=col, hue='stroke', data=df)
    plt.title(f'{col} vs. Stroke')
    plt.xticks(rotation=45)
    plt.show()
```
